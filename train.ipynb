# 市区町村名リストのダウンロード
trainデータセットの文字列を正規化するために市区町村名のデータセットを使用する

```python
import pandas as pd
import re
import kanjize

city_path = '/content/drive/My Drive/signate/市区町村名.csv'
# CSVデータをDaskで読み込む
city_data = pd.read_csv(city_path, low_memory=False)

def normalize_address(row):
    city_name = row['市区町村名']
    address = row['大字・丁目名']

    # 大字・丁目名の漢数字をアラビア数字に変換
    kanji_pattern = re.compile(r'[一二三四五六七八九十百千]+')

    def kanji_to_number(match):
        try:
            return str(kanjize.kanji2number(match.group()))
        except ValueError:
            return match.group()

    normalized_address = kanji_pattern.sub(kanji_to_number, address)
    return city_name + " " + normalized_address

# 正規化処理の適用
city_data['正規化住所'] = city_data.apply(normalize_address, axis=1)
print(city_data[['市区町村名', '大字・丁目名', '正規化住所']].head())


# 保存先のGoogle Driveフォルダのパス
folder_path = "/content/drive/My Drive/signate/city_name_files/"

# フォルダが存在しない場合は作成
os.makedirs(folder_path, exist_ok=True)

# 保存先のファイルパス
output_file_path = os.path.join(folder_path, "city_name.csv")

# データをCSVとして保存
city_data.to_csv(output_file_path, index=False)

print(f"データが保存されました: {output_file_path}")

```

# full_addressの正規化

```python
import pandas as pd
import re
import unicodedata
import kanjize
import os

# 全角・半角の統一
def normalize_text(text):
    if not isinstance(text, str):
        return ''  # テキストが文字列でない場合は空文字を返す
    return unicodedata.normalize('NFKC', text)

# 特殊文字や不要なスペースを削除する関数
def clean_text(text):
    text = text.strip()
    text = re.sub(r'\s+', ' ', text)  # 複数スペースを1つに統一
    text = re.sub(r'[^\w\s]', '', text)  # 特殊文字を削除
    return text

# 漢数字をアラビア数字に変換する関数
def kanji_to_number(match):
    kanji = match.group()
    try:
        return str(kanjize.kanji2number(kanji))
    except ValueError:
        return kanji

# full_address の正規化関数（高速化版）
def normalize_full_address_optimized(full_address, city_mapping):
    """
    full_address: 住所全体（例: 東京都千代田区一番町二丁目）
    city_mapping: 市区町村名をキー、正規化住所を値とする辞書
    """
    if not isinstance(full_address, str):
        return full_address  # 入力が文字列でない場合はそのまま返す

    # 市区町村名を辞書で検索
    for city, normalized_address in city_mapping.items():
        if full_address.startswith(city):
            rest_of_address = full_address[len(city):]
            kanji_pattern = re.compile(r'[一二三四五六七八九十百千]+')
            normalized_rest = kanji_pattern.sub(kanji_to_number, rest_of_address)
            return f"{normalized_address}{normalized_rest}"

    # 市区町村名が一致しない場合はそのまま返す
    return full_address

# データの読み込み
train_path = '/content/drive/My Drive/signate/train/train_selected.csv'
train_data = pd.read_csv(train_path, low_memory=False)

city_path = '/content/drive/My Drive/signate/city_name_files/city_name.csv'
city_data = pd.read_csv(city_path, low_memory=False)

# 市区町村名をキー、正規化住所を値とする辞書を作成
city_mapping = city_data.set_index("市区町村名")["正規化住所"].to_dict()

# full_address の正規化を適用（辞書を使用）
train_data["normalized_full_address"] = train_data["full_address"].apply(
    lambda x: normalize_full_address_optimized(x, city_mapping)
)

# 正規化後のユニークな値の数を確認
print("正規化後のユニークな値数:", train_data['normalized_full_address'].nunique())

# 正規化前と正規化後のデータの一部を表示
print("正規化前後の比較:")
print(train_data[['full_address', 'normalized_full_address']].head())

# 保存先のGoogle Driveフォルダのパス
folder_path = "/content/drive/My Drive/signate/train/"

# フォルダが存在しない場合は作成
os.makedirs(folder_path, exist_ok=True)

# 保存先のファイルパス
output_file_path = os.path.join(folder_path, "train_full_address.csv")

# データをCSVとして保存
train_data.to_csv(output_file_path, index=False)

print(f"正規化後のデータが保存されました: {output_file_path}")
```

# addr2_nameとekiの正規化処理

```python
import pandas as pd
import re
import unicodedata
import kanjize

# ファイルを読み込む
data_path = '/content/drive/My Drive/signate/train/train_full_address.csv'
train_data = pd.read_csv(data_path, low_memory=False)

# 全角・半角の統一
def normalize_text(text):
    if not isinstance(text, str):
        return ''  # テキストが文字列でない場合は空文字を返す
    return unicodedata.normalize('NFKC', text)

# 特殊文字や不要なスペースを削除する関数
def clean_text(text):
    text = text.strip()
    text = re.sub(r'\s+', ' ', text)  # 複数スペースを1つに統一
    text = re.sub(r'[\-\(\)\[\]\"\']', '', text)  # 特殊文字を削除
    return text

# 漢数字をアラビア数字に変換する関数
def kanji_to_number(match):
    kanji = match.group()
    try:
        return str(kanjize.kanji2number(kanji))
    except ValueError:
        return kanji

# addr2_name の正規化関数
def normalize_addr2_name(addr2_name):
    # 全角・半角の統一
    addr2_name = normalize_text(addr2_name)

    # 特殊文字や不要なスペースの削除
    addr2_name = clean_text(addr2_name)

    # 漢数字をアラビア数字に変換
    kanji_pattern = re.compile(r'[一二三四五六七八九十百千]+')
    addr2_name = kanji_pattern.sub(kanji_to_number, addr2_name)

    return addr2_name

# eki_name1 の正規化関数
def normalize_eki_name1(eki_name1):
    # 全角・半角の統一
    eki_name1 = normalize_text(eki_name1)

    # 特殊文字や不要なスペースの削除
    eki_name1 = clean_text(eki_name1)

    # 漢数字をアラビア数字に変換
    kanji_pattern = re.compile(r'[一二三四五六七八九十百千]+')
    eki_name1 = kanji_pattern.sub(kanji_to_number, eki_name1)

    return eki_name1


# addr2_name の正規化
train_data["normalized_addr2_name"] = train_data["addr2_name"].apply(normalize_addr2_name)

# eki_name1 の正規化
train_data["normalized_eki_name1"] = train_data["eki_name1"].apply(normalize_eki_name1)

# 正規化後のユニークな値の数を確認
print("addr2_name 正規化後のユニークな値数:", train_data['normalized_addr2_name'].nunique())
print("eki_name1 正規化後のユニークな値数:", train_data['normalized_eki_name1'].nunique())

# 正規化前と正規化後のデータの一部を表示
print("正規化前後の比較:")
print(train_data[['full_address', 'normalized_full_address', 'addr2_name', 'normalized_addr2_name', 'eki_name1', 'normalized_eki_name1']].head())

# 正規化後のデータを保存
output_path = '/content/drive/My Drive/signate/train/normalized_train_addr2_eki.csv'
train_data.to_csv(output_path, index=False)
```

```python
import pandas as pd
import numpy as np
import os

train_path = '/content/drive/My Drive/signate/train/train_post.csv'
# CSVデータを読み込む
train_data = pd.read_csv(train_path, low_memory=False, dtype={'area_JIS_code': str, 'zip_code': str})

# 列名を変更する
train_data.rename(columns={'full_address': 'land_address'}, inplace=True)
train_data.rename(columns={'addr2_name': 'sub_address'}, inplace=True)

# 標準的な建蔽率
standard_kenpei = [30, 40, 50, 60, 70, 80]

# 建蔽率の処理関数
def process_kenpei(value):
    if pd.isna(value) or value <= 0 or value > 100:  # 欠損値、0、100超えを処理
        return np.nan
    return min(standard_kenpei, key=lambda x: abs(x - value))  # 最も近い標準値

# 建蔽率の処理
train_data['land_kenpei'] = train_data['land_kenpei'].apply(process_kenpei)

# 標準的な容積率
standard_ratios = [50, 60, 80, 100, 150, 200, 300, 400, 500, 600, 700, 800, 1000]

# 容積率の処理関数
def process_youseki(x):
    if pd.isna(x) or x <= 0:  # 欠損値または0以下を処理
        return 300
    elif x == 9999:  # 異常値
        return 1000
    elif x in standard_ratios:  # 標準値をそのまま
        return x
    elif x > 1000:  # 極端な値を欠損扱い
        return np.nan
    else:  # 非標準値を丸める
        return round(x / 10) * 10

# 容積率の処理
train_data['land_youseki'] = train_data['land_youseki'].apply(process_youseki)

# land_youto と land_toshi の 0 を欠損値に置き換え
train_data['land_youto'] = train_data['land_youto'].replace(0, np.nan)
train_data['land_toshi'] = train_data['land_toshi'].replace(0, np.nan)

# 保存先のGoogle Driveフォルダのパス
folder_path = "/content/drive/My Drive/signate/train/"

# フォルダが存在しない場合は作成
os.makedirs(folder_path, exist_ok=True)

# 保存先のファイルパス
output_file_path = os.path.join(folder_path, "train_kenpei.csv")

# データをCSVとして保存
train_data.to_csv(output_file_path, index=False)

print(f"データが保存されました: {output_file_path}")

# 結果を表示
train_data.head()
```

# 築年数、year_built_quarterを追加する

```python
import pandas as pd
import numpy as np
import os

train_path = '/content/drive/My Drive/signate/train/train_kenpei.csv'

# CSVデータを読み込む
train_data = pd.read_csv(train_path, low_memory=False, dtype={'area_JIS_code': str, 'zip_code': str})


# 築年数を計算する関数
def calculate_age(target_ym, year_built):
    """
    target_ym: int (例: 202007)
    year_built: float or int (例: 200210.0)
    築年数を計算して返す
    """
    # 欠損値の場合は NaN を返す
    if pd.isna(target_ym) or pd.isna(year_built):
        return None

    # 年月形式で解釈
    target_year = target_ym // 100
    target_month = target_ym % 100
    built_year = int(year_built // 100)
    built_month = int(year_built % 100)

    # 築年数を計算
    age_years = target_year - built_year
    if target_month < built_month:  # 月が過ぎていない場合
        age_years -= 1

    return max(age_years, 0)  # 築年数が負にならないようにする

# 築年数列を作成
train_data['age_years'] = train_data.apply(lambda row: calculate_age(row['target_ym'], row['year_built']), axis=1)

# 年と月を分解
train_data['year'] = train_data['year_built'] // 100
train_data['month'] = train_data['year_built'] % 100

# 月を小数に変換して加算（12月の場合を考慮）
train_data['year_built_quarter'] = train_data['year'] + (
    (train_data['month'] - 1) / 12
)

# zip_code を文字列型に変換
train_data['zip_code'] = train_data['zip_code'].astype(str)

# 保存先のGoogle Driveフォルダのパス
folder_path = "/content/drive/My Drive/signate/train/"

# フォルダが存在しない場合は作成
os.makedirs(folder_path, exist_ok=True)

# 保存先のファイルパス
output_file_path = os.path.join(folder_path, "train_built_zip.csv")

# データをCSVとして保存
train_data.to_csv(output_file_path, index=False)

print(f"データが保存されました: {output_file_path}")

# 結果の確認
print(train_data[['year_built', 'age_years', 'year_built_quarter', 'zip_code']].head())
```

# target_ym を四半期形式に変換

```python
import pandas as pd
import numpy as np
import os

train_path = '/content/drive/My Drive/signate/train/train_built_zip.csv'

# CSVデータを読み込む
train_data = pd.read_csv(train_path, low_memory=False, dtype={'area_JIS_code': str, 'zip_code': str})


# 月ごとの四半期マッピング
quarter_mapping = {
    1: 0.25,  # Q1
    7: 0.75   # Q3
}

# target_ym を四半期形式に変換
def convert_to_quarter(target_ym):
    year = target_ym // 100  # 年部分を抽出
    month = target_ym % 100  # 月部分を抽出
    quarter = quarter_mapping.get(month, 0)  # 四半期にマッピング
    return year + quarter

train_data["target_quarter"] = train_data["target_ym"].apply(convert_to_quarter)

# 保存先のGoogle Driveフォルダのパス
folder_path = "/content/drive/My Drive/signate/train/"

# フォルダが存在しない場合は作成
os.makedirs(folder_path, exist_ok=True)

# 保存先のファイルパス
output_file_path = os.path.join(folder_path, "train_target_quarter.csv")

# データをCSVとして保存
train_data.to_csv(output_file_path, index=False)

print(f"データが保存されました: {output_file_path}")

train_data.head()
```
